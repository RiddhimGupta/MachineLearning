```
#split
#mydata.split() #split all string by \n new line
import time
```

```
newdata1=[i for i in mydata.split()]
```

```
#NLTK graphs
import matplotlib.pyplot as plt
import nltk

%matplotlib inline #incase matplotlib is not supporting by default with pandas or nltk
#to replacd plt.show() from any platform
```

```
dir(nltk)
```

```
[i for i in dir(nltk) if 'Freq' in i] #this will plot for
```

```
nlpdata=nltk.FreqDist(newdata1)
```

```
nlpdata.plot(20)
```

```
#we can remove stop words from newdata1 then plot the graph
from nltk.corpus import stopwords
```

```
#removing stopwords
removedata=[i for i in newdata1 if i not in stopwords.words('english')]
```

```
#removedata
#apply NLTK FreqDist
nlpremove=nltk.FreqDist(removedata)
```

```
nlpremove.plot(20)
```
